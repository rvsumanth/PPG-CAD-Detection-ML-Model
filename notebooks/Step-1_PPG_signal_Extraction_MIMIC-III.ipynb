```python
BASE_ROOT = "mimic3wdb-matched/1.0"
GROUP_NAME = "p08"   # change to "p02" when needed
```

##**Function for Extracting Quality PPG**

```python
def extract_clean_ppg_min3_to_8(record_name, pn_dir):
    try:
        header = wfdb.rdheader(record_name, pn_dir=pn_dir)
        fs = int(header.fs)

        samples_per_min = fs * 60
        skip_samples = 2 * samples_per_min
        keep_samples = 6 * samples_per_min
        total_samples = skip_samples + keep_samples

        # find PPG channel
        ppg_index = None
        for i, sig in enumerate(header.sig_name):
            if "PLETH" in sig.upper() or "PPG" in sig.upper():
                ppg_index = i
                break

        if ppg_index is None:
            return None

        signal, _ = wfdb.rdsamp(
            record_name,
            pn_dir=pn_dir,
            channels=[ppg_index],
            sampto=total_samples
        )

        ppg = signal[:, 0]
        ppg = ppg[np.isfinite(ppg)]

        if len(ppg) < total_samples:
            return None

        # keep minute 3 → 8
        ppg = ppg[skip_samples: skip_samples + keep_samples]

        # quality checks
        if np.std(ppg) < 0.01:
            return None
        if np.ptp(ppg) < 0.05:
            return None

        return ppg, fs

    except:
        return None
```

##**Extracting each and every Patient's PPG data from Records**

```python
rows = []

saved = 0
skipped = 0

print(f"\nProcessing folder {GROUP_NAME}")

for patient_id in patient_list:
    pn_dir = f"{BASE_ROOT}/{GROUP_NAME}/{patient_id}"

    try:
        records = wfdb.get_record_list(pn_dir)
    except:
        print(f"Cannot access {patient_id}")
        skipped += 1
        continue

    print(f"\nPatient {patient_id} | Records: {len(records)}")

    patient_saved = False

    for rec in records:
        result = extract_clean_ppg_min3_to_8(rec, pn_dir)
        if result is None:
            continue

        ppg, fs = result
        samples_per_min = fs * 60

        row = {
            "patient_id": patient_id,
            "group": GROUP_NAME,
            "record_id": rec,
            "fs": fs
        }

        for m in range(6):
            start = m * samples_per_min
            end = start + samples_per_min
            row[f"ppg_min_{m+3}"] = " ".join(
                f"{x:.4f}" for x in ppg[start:end]
            )

        rows.append(row)
        saved += 1
        patient_saved = True

        print(f"Saved PPG | {patient_id} | Record: {rec}")
        break   # ONE PPG PER PATIENT

    if not patient_saved:
        print(f"No valid PPG for {patient_id}")
        skipped += 1
```

##**Saving the Extracted PPG data into CSV folder**

```python
df = pd.DataFrame(rows)
out = '/content/drive/MyDrive/Colab Notebooks/Final Project/ds'
OUTPUT = os.path.join(out, "ppg-p08-data.csv")
df.to_csv(OUTPUT, index=False)

print("SUMMARY")
print("Folder       :", GROUP_NAME)
print("Patients in  :", len(patient_list))
print("Saved        :", saved)
print("Skipped      :", skipped)
print("CSV saved")
```
