#**Merging the Multiple patients data**

```python
import pandas as pd
import os
from pathlib import Path
```

```python
DATA_DIR = Path('/content/drive/MyDrive/Colab Notebooks/Final Project/ALL DATA SETS/PPG Data')
```

```python
all_dfs = []
reference_columns = None

for csv_file in sorted(DATA_DIR.glob("*.csv")):
    print(f"Reading {csv_file.name}")

    df = pd.read_csv(csv_file)

    # Column validation (CRITICAL)
    if reference_columns is None:
        reference_columns = df.columns.tolist()
    else:
        if df.columns.tolist() != reference_columns:
            raise ValueError(
                f"Column mismatch in {csv_file.name}\n"
                f"Expected: {reference_columns}\n"
                f"Found:    {df.columns.tolist()}"
            )

    all_dfs.append(df)
```

```python
# Merge all data
master_df = pd.concat(all_dfs, ignore_index=True)
```

```python
# Save final dataset
out = '/content/drive/MyDrive/Colab Notebooks/Final Project/ds'
output = os.path.join(out, "Master-ppg-data.csv")
master_df.to_csv(output, index=False)

print("\nMerge completed successfully")
print("Total rows:", len(master_df))
print("Total files merged:", len(all_dfs))
```
